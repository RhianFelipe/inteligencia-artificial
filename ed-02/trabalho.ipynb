```python
lista_fat = []
def fatorial(n):
    fat = 1
    i = 1
    while i <= n:
        fat = fat*i
        lista_fat.append(fat)
        print(f"O valor de {i}! eh = {lista_fat}") 
        i = i + 1
        
#-----
fatorial(10)
lista_fat.sort(reverse=True)
print(f'Lista Fatorial decrescente:{lista_fat}')

```
